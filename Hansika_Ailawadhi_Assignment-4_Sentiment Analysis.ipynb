### 5. Using TextBlob for Sentiment Analysis

```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
```

```python
df['sentiment'] = df['description'].apply(get_sentiment)
```

### 6. Plotting Sentiment Distribution

```python
import matplotlib.pyplot as plt

plt.hist(df['sentiment'], bins=20, color='blue', alpha=0.7)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculating average sentiment polarity for each type
average_sentiment_by_type = df.groupby('genres')['sentiment'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='genres', y='sentiment', data=average_sentiment_by_type, palette='viridis')
plt.title('Average Sentiment by genres')
plt.xlabel('Genres')
plt.ylabel('Average Sentiment Polarity')
plt.xticks(rotation=45)
plt.show()
```
